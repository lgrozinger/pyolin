To import the gates from a CSV pandas dataframe

```python
import pandas
raw = pandas.read_csv('../cyto2func/standardised_cheeky.csv')
renamed = raw.rename(columns={'rrpu': 'decomp_flor'}).rename(columns={'newstandard': 'rrpu'})

from pyolin.dataframe import GateData
data = GateData(renamed)
```

Select a gate, or ranges of gates, using Python array slicing syntax

```python
gate = data['DH5alpha':'pAN':'Amer_f1']
gate.rpuplot();
```

Log transformed output of gates

```python
log_gate = gate.log()
log_gate.plot(ylimits=(1e-1, 1e1));
```

As is normalisation

```python
normal_gate = gate.normal(lower=(1.0, 1.0), upper=(2.0, 2.0))
normal_gate.rpuplot();
```

Transforms can be combined

```python
log_normal_gate = gate.log().normal(lower=(1.0, 1.0), upper=(2.0, 2.0))
log_normal_gate.rpuplot();
```
